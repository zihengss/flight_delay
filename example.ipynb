```python
import sqlite3

conn = sqlite3.connect("flight_delay_analysis.db")
cur = conn.cursor()

# List all tables
cur.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = [t[0] for t in cur.fetchall()]

for t in tables:
    print(f"\n===== {t} =====")
    cur.execute(f"PRAGMA table_info({t});")
    for col in cur.fetchall():
        print(col)
```

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("flight_delay_analysis.db")

query = """
SELECT *
FROM Airline A
JOIN Weather W
    ON A.FL_DATE=W.DATE AND A.ORIGIN_CITY_NAME=W.CITY_NAME and A.ORIGIN_STATE_ABR=W.STATE_ABBR
JOIN FuelPrice FP
    ON A.FL_DATE=FP.DATE
JOIN StockPrice SP
    ON A.FL_DATE=SP.DATE AND A.OP_UNIQUE_CARRIER=SP.iata_code
WHERE A.OP_UNIQUE_CARRIER='AA'
"""

df = pd.read_sql_query(query, conn)
df.head()
```
